```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
%matplotlib inline

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedShuffleSplit, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer, OneHotEncoder
from sklearn.metrics import auc, roc_auc_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
```

## Function List

```python
names = ["LogisticRegression","Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
             "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
             "Naive Bayes", "Gradient Boosting Classifer", "XGBoost"]

classifiers = [
        LogisticRegression(),
        KNeighborsClassifier(3),
        SVC(kernel="linear", C=0.025),
        SVC(gamma=2, C=1),
        GaussianProcessClassifier(1.0 * RBF(1.0)),
        DecisionTreeClassifier(max_depth=5),
        RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        MLPClassifier(alpha=1,max_iter=500),
        AdaBoostClassifier(),
        GaussianNB(),
        GradientBoostingClassifier(max_depth=5),
        xgb.XGBClassifier()]

onehot = OneHotEncoder()
lr = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
# Create list for object and numeric features

def feature_list(df):
    obj_features = []
    num_features = []

    for i in df.columns:
        if df[i].dtype == "object":
            obj_features.append(i)
        elif i == 'poi':
            pass
        else:
            num_features.append(i)
    return obj_features, num_features

# Calculate CV score for each classifer

def cv_score(df, X_val=None):
#     obj_features, num_features = feature_list(df)
    names = ["LogisticRegression","Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
             "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
             "Naive Bayes", "Gradient Boosting Classifer", "XGBoost"]

    classifiers = [
        LogisticRegression(),
        KNeighborsClassifier(3),
        SVC(kernel="linear", C=0.025),
        SVC(gamma=2, C=1),
        GaussianProcessClassifier(1.0 * RBF(1.0)),
        DecisionTreeClassifier(max_depth=5),
        RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        MLPClassifier(alpha=1,max_iter=500),
        AdaBoostClassifier(),
        GaussianNB(),
        GradientBoostingClassifier(max_depth=5),
        xgb.XGBClassifier()]
    print("==========================")
    print("Cross Validation")
    print("==========================")

    for name, clf in zip(names, classifiers):
        cv_score = cross_val_score(clf, df[num_features], df['poi'],cv=5, scoring='roc_auc').mean()
        print(f'{name}: {cv_score}')

    print()
    print("====================")
    print()
    if X_val is None:
        pass
    else:
        print("==========================")
        print("ROC_AUC_SCORE using validation set")
        print("==========================")
        for name, clf in zip(names, classifiers):
            clf.fit(df[num_features],df['poi'])
            y_pred = clf.predict(X_val[num_features])
            roc_score = roc_auc_score(X_val['poi'], y_pred)
            print(f'{name}: {roc_score}')
            
            
# Training set Fill NA and Split
def train_fillna_split(train):
    train = train.fillna(0)
    split = StratifiedShuffleSplit(n_splits=1, test_size =0.2, random_state=42)
    for train_index, val_index in split.split(train, train['poi']):
        X_train = train.loc[train_index]
        X_val = train.loc[val_index]
        y_train = X_train['poi']
        y_val = X_val['poi']
    return X_train, y_train, X_val, y_val

def draw_kde(df):
    plt.figure(figsize = (40, 120))

    num = len(df.columns)
    for i, feature in enumerate(df.columns):
        plt.subplot(10,2,i+1)

        sns.kdeplot(df.loc[df['poi']==0,feature], label="poi ==0")
        plt.xlabel(feature,size= 20)
        plt.ylabel("density", size= 20)
        sns.kdeplot(df.loc[df['poi']==1,feature], label="poi ==1")
#         plt.show()
    
#     sns.kdeplot(train.loc[train['poi'] == 0, 'to_messages'],label='target ==0')
#     sns.kdeplot(train.loc[train['poi'] == 1, 'to_messages'],label='target ==1')
#     plt.show()
```
